```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier, LogisticRegression, SGDRegressor, LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import plotly.graph_objects as go
```

# DIABETES

```python
dfDiabetes = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv')
dfDiabetes.head(30)
```

```python
dfDiabetes.iloc[:,-1].head(30)
```

```python
procesito = MinMaxScaler()
procesito.fit(dfDiabetes.iloc[:,:-1])
dfDiabetes.iloc[:,:-1] = procesito.transform(dfDiabetes.iloc[:,:-1])
dfDiabetes
```

```python
X_Diabetes = dfDiabetes.iloc[:,:-1]
y_Diabetes = dfDiabetes.iloc[:,-1]
```

```python
X_trainDiabetes, X_testDiabetes, y_trainDiabetes, y_testDiabetes = train_test_split(X_Diabetes, y_Diabetes, test_size=0.2, random_state=0)
y_testDiabetes
```

## MANUAL

```python
def cost_function(y, y_pred):
    n = len(y)
    cost = (1/(2*n))*np.sum((y_pred-y)**2)
    return cost

def predict(x, w, b):
    return np.dot(x, w) + b

def predict_proba(x, w, b):
    return 1/(1+np.exp(-predict(x, w, b)))
 
def get_gradient(x, y, y_proba, w, b):
    n = len(y)
    dw = (1/n)*np.dot(x.T, (y_proba-y))
    db = (1/n)*np.sum(y_proba-y)
    return dw, db

def update_params(w, b, dw, db, learning_rate):
    w = w - learning_rate*dw
    b = b - learning_rate*db
    return w, b

def reg_log(X_train, y_train, niter=100, learning_rate=0.1):
    w = np.zeros(X_train.shape[1])
    b = 0
    for i in range(niter):
        y_proba = predict_proba(X_train, w, b)
        dw, db = get_gradient(X_train, y_train, y_proba, w, b)
        w, b = update_params(w, b, dw, db, learning_rate)
        if i%100==0:
            print(f'Costo: {cost_function(y_train, y_proba)}')
    return w, b
```

```python
#Calcular el acurracy en train y test sin funciones
w, b = reg_log(X_trainDiabetes, y_trainDiabetes)
y_pred_train = predict_proba(X_trainDiabetes, w, b)
y_pred_test = predict_proba(X_testDiabetes, w, b)
y_pred_train = np.where(y_pred_train>=0.5, 1, 0)
y_pred_test = np.where(y_pred_test>=0.5, 1, 0)
print(f'Accuracy train: {np.mean(y_pred_train==y_trainDiabetes)}')
print(f'Accuracy test: {np.mean(y_pred_test==y_testDiabetes)}')
```

# CALIFORNIA

```python
dfCaliforniaTrain = pd.read_csv('https://raw.githubusercontent.com/DanielOlarte-GitHub/DataSetsFamosos/main/california_housing_train.csv')
dfCaliforniaTest = pd.read_csv('https://raw.githubusercontent.com/DanielOlarte-GitHub/DataSetsFamosos/main/california_housing_test.csv')
dfCaliforniaTrain.shape, dfCaliforniaTest.shape
```

```python
dfCaliforniaTest.head()
```

```python
X_dfCaliforniaTest = dfCaliforniaTest.iloc[:,:-1]
y_dfCaliforniaTest = dfCaliforniaTest.iloc[:,-1]
X_dfCaliforniaTrain = dfCaliforniaTrain.iloc[:,:-1]
y_dfCaliforniaTrain = dfCaliforniaTrain.iloc[:,-1]
y_dfCaliforniaTrain
```

```python
X_dfCaliforniaTest.head(2)
```

```python
escaladitoCalifornia = MinMaxScaler()
escaladitoCalifornia.fit(X_dfCaliforniaTrain)
X_dfCaliforniaTrain = escaladitoCalifornia.transform(X_dfCaliforniaTrain)
X_dfCaliforniaTest = escaladitoCalifornia.transform(X_dfCaliforniaTest)
```

```python
X_dfCaliforniaTest
```

```python
y_dfCaliforniaTest
```

```python

def cost_function(y, y_proba):
    y_proba = np.clip(y_proba, 1e-10, 1-1e-10) # evitar valores muy cercanos a 0 o 1
    cost = -np.mean(y*np.log(y_proba)+(1-y)*np.log(1-y_proba))
    return cost


def predict(X, w, b):
    return np.dot(X, w) + b

def get_gradient(X, y, y_proba, w, b):
    n = len(y)
    dw = (1/n)*np.dot(X.T, (y_proba-y))
    db = (1/n)*np.sum(y_proba-y)
    return dw, db

def update_params(w, b, dw, db, learning_rate= 0.1):
    w = w - learning_rate*dw
    b = b - learning_rate*db
    return w, b

def reg_log(X, y, niter=100, learning_rate=0.1):
    w = np.random.randn(X.shape[1])
    b = 0
    for i in range(niter):
        y_proba = predict(X, w, b)
        dw, db = get_gradient(X, y, y_proba, w, b)
        w, b = update_params(w, b, dw, db, learning_rate)
        if i%100==0:
            print(f'Costo: {cost_function(y, y_proba)}')
    return w, b
```

```python
# Calcular mse en train y test. con las funciones
w, b = reg_log(X_dfCaliforniaTrain, y_dfCaliforniaTrain)
y_pred_train = predict(X_dfCaliforniaTrain, w, b)
y_pred_test = predict(X_dfCaliforniaTest, w, b)
print(f'MSE train: {mean_squared_error(y_dfCaliforniaTrain, y_pred_train, squared=False)}')
print(f'MSE test: {mean_squared_error(y_dfCaliforniaTest, y_pred_test, squared=False)}')
```

## SGD Regressor

```python
modeloCaliSGD = SGDRegressor()
modeloCaliSGD.fit(X_dfCaliforniaTrain, y_dfCaliforniaTrain)
y_pred_train = modeloCaliSGD.predict(X_dfCaliforniaTrain)
y_pred_test = modeloCaliSGD.predict(X_dfCaliforniaTest)
print(f'MSE train: {mean_squared_error(y_dfCaliforniaTrain, y_pred_train, squared=False)}')
print(f'MSE test: {mean_squared_error(y_dfCaliforniaTest, y_pred_test, squared=False)}')
```

## LINEAR REGRESSION

```python
modeloCaliReg = LinearRegression()
modeloCaliReg.fit(X_dfCaliforniaTrain, y_dfCaliforniaTrain)
y_pred_train = modeloCaliReg.predict(X_dfCaliforniaTrain)
y_pred_test = modeloCaliReg.predict(X_dfCaliforniaTest)
print(f'MSE train: {mean_squared_error(y_dfCaliforniaTrain, y_pred_train, squared=False)}')
print(f'MSE test: {mean_squared_error(y_dfCaliforniaTest, y_pred_test, squared=False)}')
```
